```python
# creating new features based on existing columns/features
# create input relationship -> inc efficiency , accuracy , performance
df1
```

```python
# create total_income of ApplicantIncome + CoapplicantIncome
df1["total_income"] = df1["ApplicantIncome"]+df1["CoapplicantIncome"] # at the end 
```

```python
df1
```

```python
# loan to income ratio -> total_income how much starting loan i will grant for customer 
# ApplicantIncome more then small amount of loan grant is easy 
# division between loan_amount and total_income
# LTI -> Loan to income ratio

df1["LTI"] = (df1["LoanAmount"]/df1["total_income"])*100
df1
```

```python
# create the new feature by binning them into a group 
# binning -> grouping 
# total_income as < 5000 -> low income 
# ti > 5000 and ti < 10000: avg
# ti > 10000: high income 

def income_group(var):
    if var < 5000:
        return "Low income"
    elif var > 5000 and var < 10000:
        return "Mid income"
    else:
        return "High Income"
```

```python
income_group(500)
```

```python
df1["binning"] = df1["total_income"].apply(income_group)
df1
```

```python
# create column for stability and unsatbility
# married, edu: graduate & not self-employed -----> stable
# else unstable

def stable(var):
    if (var["Married"] == "Yes") & (var["Education"] == "Graduate") & (var["Self_Employed"] == "No"):
        return "satble"
    else:
        return "unstable"
```

```python
df1["satbility"] = df1.apply(stable, axis=1)
df1
```

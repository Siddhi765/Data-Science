# Array Functions

```python
# prod -> To do product of elements present in the array 
from numpy.core.fromnumeric import prod
print(np.prod(a2)) # becz it contain zeroo -> any thing multiply by zero is zero 
```

```python
print(np.prod(a4))
```

```python
# Minimum and maximum
print("Minimum:" , np.min(a2))
print("Minimum:" , np.min(a4))
print("Maximum:" , np.max(a2))
print("Maximum:" , np.max(a4))
```

```python
# sum and mean -> average
print("sum: " , np.sum(a2))
print("mean:", np.mean(a2))
```

```python
# finding mean across rows(axis=1) and column(axis=0)
print("mean across rows: ", np.mean(a2,axis=1))
print("mean across columns: ", np.mean(a2,axis=0))
```

```python
# standard deviation -> it the measure of how much variation exists from the mean of a data
print("\"S.D:\"", np.std(a2))
# variance
print("\"Variance\":", np.var(a2))
```

# Matrix Operations

```python
# dot product of two matrix -> multiplication scalar values
a5 = np.arange(15).reshape(3,5)
a6 = np.arange(15).reshape(5,3)
result = np.dot(a5,a6)
print(result)
```

```python
# how to iterate over the matrix 
# for loop -> 2D in nature , uses nested list
mat = a5 = np.arange(4).reshape(2,2)
for i in mat:
    print(i)
```

```python
# Iterating Each Element (Nested Loop)
# If you want to access every element, use a nested loop.

for i in mat:
    for j in i:
        print(j)
```

```python
# Using NumPy Iterator (nditer) -> iterate in nested loop
# NumPy also provides a special iterator.

for element in np.nditer(mat):
    print(element)
```

```python
# how to convert 1D to 2D -> reshape
# how to convert 2D to 1D -> ravel
x = mat.ravel() # 2D -> 1D
print(x)
```

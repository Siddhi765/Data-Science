# Part 5

# Empirical Rule
![My Image](empirical_rule.png)

```python
# what is the expected range of return for 95% confidence interval ?
# 1. which statistical parameter
# 2. which column 
# 3. z-score (1.96)

# use empirical rule for calculation 

# alpha = interval/confidence level 95% ---> 0.95 
# loc = location of parameter distribution,mean for distribution
# scale = scale of parameter distribution,std for distribution

from scipy import stats

df1_ci = stats.norm.interval(0.95,loc=df1["gain"].mean(),scale=df1["gain"].std())
print(df1_ci) 
# z-value -> -2.5% (loss) and +2.6% (gain)
```

```python
df2_ci = stats.norm.interval(0.95,loc=df2["gain"].mean(),scale=df2["gain"].std())
df2_ci
# z-value -> -5.1% (loss) and +5.2% (gain)
```

```python
passport = pd.DataFrame({"Processing_time":[16, 16, 30, 37, 25, 22, 19, 35, 27, 32, 34, 28, 24, 35, 24, 21, 32, 29, 24, 35, 28, 29, 18, 31, 28, 33, 32, 34, 25, 22, 21, 27, 41, 23, 23, 16, 24, 38, 26, 28]})
```

passport office claims that passport applications are 
processedd within 30 days submitting application on an avg
sample of data as 40
we have to conduct hypothesis test at significance level a = 0.95 and the population std deviation
given is 12.5 days

Formula for z-test
$$
Z = \frac{\bar{X} - \mu}{\sigma/\sqrt{n}}
$$

when to use:
1. variance of population is known
2. N > 30
3. population mean is given
4. follows normal distribution

![My Image](distribution.png)

```python
# H0 : 𝜇 >= 30
# H1 : 𝜇 < 30
# use z-table for stats.norm.cdf(z_score) -> P-Value

from scipy import stats
import math
def z_test(pop_mean, pop_std, sample):
    z_score = (sample.mean()-pop_mean)/(pop_std/math.sqrt(len(sample)))
    return z_score, stats.norm.cdf(z_score)
```

```python
z_test(30,12.5,passport["Processing_time"])
# p_value < alpha -> reject H0
# p_value > alpha -> accept H0
```

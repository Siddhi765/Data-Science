# Plotting Part 2

# Univariate numerical data

```python
# distribution graph
plt.figure(figsize = (10,7)) # To enhance the size of graph
sns.distplot(df1["LoanAmount"])
```

```python
plt.figure(figsize=(11,11))
sns.histplot(df1["LoanAmount"])
```

# Bivariate numerical data


```python
# two features at a time 
# joint and clot
# axisgrid -> to point accurate coordinate thats why resizeing is not allowed
# hue -> to judge the coordinate -> to read the coordinate 
# hue="Dependents" -> to find outliers -> 3+ represent outliers
# kind="scatter" -> to get insights 

sns.jointplot(data=df1,x="ApplicantIncome", y="LoanAmount", kind="scatter", hue="Dependents")
```

```python
# Scatter Plotting for the combination of every numerical feature along with the every entry
sns.pairplot(data=df1)
```

# Line gragh

```python
x=[5,2,5,7,9]
y=[10,5,8,4,5]

# inc figure size
plt.figure(figsize=(10,5))

# marker -> to put the coordinate 
plt.plot(x,y,marker="*") # * -> for checkpoints

# color 
# plt.plot(x,y,"o:r") # to change the color -> "marker : color"
# plt.plot(x,y,"o-.r") # to change the line of graph dashed , dotted -> "marker-or-.color"

# labeling of x and y axis

plt.xlabel("Profit")
plt.ylabel("Sales")
plt.title("Sales Analysis")
```

# Stacked plots

```python
x_1 = ["A","B","C","D"]
y_1 = [10,20,10,30]
y_2 = [20,25,15,25]

plt.bar(x_1,y_1,color = "k")

# stacking up on to one another 
plt.bar(x_1,y_2,bottom=y_1,color="r")
```

```python
# to create circle and mention the coordinate , radius , face color , where to draw the circle

var = plt.Circle((0,0),radius=0.5,fc="r")
plt.gca().add_patch(var)
plt.axis("scaled")
```

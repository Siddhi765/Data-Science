# List Comprehension

Q.1 square the given number present inside the list ande gfive squared number output in the list format

```python
list_1 = [10,20,304,50,60]
list_output = []
for i in list_1:
    z = i ** 2
    list_output.append(z)
    print(list_output)
```

## List Comprehension without conditional statements

```python
# List Comprehension
# syntax : [output for loop]

# example
# print n = 10 natural numbers inside the list
list_natural = [i for i in range(10)]
print(list_natural)
```

## List Comprehension with conditional statements -> if without else

```python
# syntax 
# var_1 = [output for i in range(n) if condition]

# e.g 
# pick up n = 12 natural numbers and print only even numbers inside the list
list_var = [i for i in range(12) if i % 2 == 0]
print(list_var)
```

## List Comprehension with conditional statements -> if with else

```python
# syntax 
# var_1 = [output if condition else output for i in range(n)]

# e.g 
# print "even" in place of even numbers and "odd" in place of odd numbers
list_var = ["even" if i % 2 == 0 else "odd" for i in range(1,12)] # in range(1,12) it does not print for '0'
print(list_var)
```

# Questions on list comprehension

```python
# check if last alphabet of the string present inside the list is "a" or not 
# if "a" print "yes"
# if not ending with "a" print string as it is

list_str = ["banana","guava","apple","dragon fruit"]
practice = ["yes" if i[-1] == "a" else i for i in list_str] # i[-1] -> extract the last charactor of the str -> banana -> a
print(practice)
```

```python
# you are having two list 
# perform the product of respective index position
#  use zip() fuction

list_1 = [2,3,4,5,6]
list_2 = [2,4,10,5,8]

# Normal programming 
list_product = []
for i,j in zip(list_1,list_2):
    list_product.append(i*j)
print(list_product)
```

```python
# by using list comprehenssion
product_output = [i*j for i,j in zip(list_1,list_2)]
print(product_output)
```

# Dict comprehension

```python
# without comprehension

list_1 = [2,3,4,5,6]

# dict_1 = {2:4,3:9,4:16,5:25,6:36} # output dictionary
dict_emp = {}
for i in list_1:
    dict_emp[i] = i ** 2
print(dict_emp)
```

```python
# with comprehension
# syntax
# var_2 = {i:output for i in iterable}

list_1 = [2,3,4,5,6]
dict_var_2 = {i:i**2 for i in list_1 }
print(dict_var_2)
```

## Dict Comprehension with conditional statements -> if without else

```python
# syntax
# var_2 = {i:output for i in iterable if condition}

# Q.1 given the list print the word of the list as "key" of the dict 
# the last alphabet of the word as "value" inside the list 
list_given = ["Badminton","basketball","kho-kho","Dancing"]

dict_output = {i:i[-1] for i in list_given}
print(dict_output)
```

```python
# Q.2 pick up "n" natural number and place inside a list
# after that find the number which are less than 5 and perform the cube of that number
# and place the number as "key" in dict and cube of number as "value" in dict

n = int(input("Enter the value of n : "))

list_natural = [i for i in range(n+1)]
print(list_natural)

dict_output = {i:i**3 for i in list_natural if i < 5}
print(dict_output)
```

## Dict Comprehension with conditional statements -> if with else

```python
# syntax
# var_2 = {output_true if condition else output_false i:output for i in iterable}
nums = [1,2,3,4]

square = {i:i*i for i in nums}

print(square)
```

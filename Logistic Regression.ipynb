## Assumption of Logistic Regression

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Siddhi\Desktop\Data Science\Dataset\Social_Network_Ads.csv")
df
```

```python
df.dtypes
```

```python
x=df.iloc[:,[2,3]]
y=df.iloc[:,4]
```

```python
#splitting the data:

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=4)
```

## Feature Scaling

### Standardization
### Normalization

```python
from sklearn.preprocessing import StandardScaler

var = StandardScaler()
x_train = var.fit_transform(x_train)
x_test = var.transform(x_test)
```

```python
x_train
```

## Fitting of Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

class_ = LogisticRegression()
class_.fit(x_train, y_train)
```

```python
y_pred = class_.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

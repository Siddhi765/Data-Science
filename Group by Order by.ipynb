```python
# group by on the basis of two column
var = df.groupby(["continent","country"]) # write inside the list
var.first()
```

```python
'''
Aggregate function takes part 
always apply groupby to discrete type of values 
continuous values -> aggregate function
5 types
    sum
    min
    max
    count
    avg
'''
```

```python
df.dtypes
```

```python
# what is the total wine surving for the asia continent

# groupby on continent -> use aggregate function -> .agg({"column_to_apply" : "function_to_apply"})

df.groupby("continent").agg({"wine_servings" : "sum"})
```

```python
# to validate it 
(x.get_group("Asia")).sum()
```

```python
# perform sum of every entry
df.groupby("continent").sum()
```

```python
# sum and average of wine_servings together for "Asia" continent

df.groupby("continent").agg({"wine_servings" : ["sum","mean"],"spirit_servings":"min"})
```

# KNN (K Nearest Neighbour)

Types Algorithm to fine value of k in KNN:

1. Eucledian 
$$
d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}
$$

2. Manhattan
$$
d = |x_1 - x_2| + |y_1 - y_2|
$$

### Example of KNN
![My Image](KNN_example.png)

Eucledian Formula for k
$$
d = \sqrt{\sum_{i=1}^{k} (x_i - y_i)^2}
$$

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Siddhi\Desktop\Data Science\Dataset\Social_Network_Ads.csv")
df
```

```python
# Input : Age, EstimatedSalary
# Output : Purchased

x = df.iloc[:,2:4]
y = df.iloc[:,[-1]]
x
```

```python
y
```

```python
# split the dataset into train-test-split

from sklearn.model_selection import train_test_split # for spliting the data

# dont shuffle the position
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 4)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
x_train # random_state -> random selecttion of rows
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier

Knn = KNeighborsClassifier(n_neighbors = 17)
Knn.fit(x_train, y_train)
y_pred = Knn.predict(x_test)
```

# Confusion Matrix

![My Image](confusion_matrix.png)

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))

# Error-rate
print(1-accuracy_score(y_test,y_pred))
```

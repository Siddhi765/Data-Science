```python
# how many female got the loan 
# crosstab -> crosstable -> table with two categorical data 
# margins=True -> overall entries
pd.crosstab(df1["Loan_Status"],df1["Gender"],margins=True)
```

```python
# To calculate in percentage -> acrross rows
pd.crosstab(df1["Loan_Status"],df1["Gender"],normalize="index")*100
```

```python
# To calculate in percentage -> acrross columns
pd.crosstab(df1["Loan_Status"],df1["Gender"],normalize="columns")*100
```

```python
df1
```

```python
# find the average loan amount for the female and max applicant income for female

w = df1.groupby(["Gender"]).agg({"LoanAmount" : "mean" , "ApplicantIncome" : "max"})
w
```

```python
# Filtration
w = w.reset_index()
w
```

```python
# iloc -> for reseting index 
w = w.iloc[0:1,:]
w
```

```python
# renaming of the column

w = w.rename(columns = {"LoanAmount" : "Mean_LoanAmount" , "ApplicantIncome" : "max_ApplicantIncome"})
w
```

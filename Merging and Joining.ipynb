```python
# Merging

# avoid duplicate entry
# uses Joins -> joins two datasets 
# types :-  left -> column 1 with common entry
#           right -> column 2 with common entry
#           inner -> only common entry 
#           outer -> common entry only one time with all entry of column 1 and column 2
# common entry -> sangeeta
# outer is suitable for us 
```

```python
merged_df = pd.merge(div_3, div_4, on=["roll_no","First_name","Last_name"], how="outer")
merged_df
```

# keyword Identifier

## keyword

reserved word. 36 keywords . e.g def , import, True, False -> they carry some meaning

## identifier

those words which helps user to assign specific meaning to it & these are not reserved
rule:
1. cannot be keywords
2. cannot contain any special char like @#$!
3. always start with _ of alphabet & then followed by any combination of charac & number 0-9
e.g Megha_87 = 67 , _key = 87

```python
Megha_08 = 67
```

```python
Megha_123 = 90
```

```python
_sid316_06 = 0 # identifiers
```

```python
print(list_1[3])
```

```python
# nested list 
# list inside list 
list_nested =      [10,20,30,40,["apple","mango","guava","banana","grapes"],89]
# indexing_outer -> 0  1  2   3  ------------------4----------------------- 5
# indexing_inner ->               0        1       2        3       4
print(list_nested)
print(list_nested[4]) # inner list 
print(list_nested[4][1]) # inner list ka extract element
print(list_nested[5]) # inner list 
print(list_nested[-2][2]) # inner list 
```

# Methos for list

```python
# methods 
# dedicated for datatypes
# start with . and end with ()
# some methods are used only for list
# make changes permanentely
```

```python
# append
# add element at the end of the list when list is already declared
# .append("element to be append")
list_my = [1,2,3,4,5,"apple",7,89,3.4]
list_my.append("MANGO")
print(list_my)
```

```python
# extend 
# add multiple element at end of the list
list_1 = [ 1,20.2,"String",True,None,14]
list_1.extend(["banana","litchi"])
print(list_1)
```

```python
# insert
# add element at specific position
# .insert(index_position , element_to_be_inserted)
list_1.insert(3,"new")
```

```python
print(list_1)
```

```python
# delete element from the end of the list
list_del = [1, 20.2, 'String', True, None, 14, 'banana', 'litchi']
list_del.pop()
print(list_del)
```

```python
# delete element from specific index position
list_del.pop(2)
print(list_del)
```

```python
# remove 
# delete element from yhe list 
list_del.remove("banana")
print(list_del)
```

```python
list_large = ["apple","banana",1,2,3,4,5,6,7,8,9.6,5.6,"xyz"]
print(list_large)
# create a copy of all elements present inside list of large
# stored in separate memory location independently
copy_list = list_large.copy()
print(copy_list)
```

```python
# sort 
# to sort list in ascending or descending order 
list_sort = [1,2,4,7,3,100.600,300,12,21,45,41]
list_sort.sort() # ascending order
print(list_sort)
list_sort.sort(reverse=True) # descending order
print(list_sort)
```

```python
# reverse 
list_rev = [1,3,6,2.4,"apple","banana",78,89,"xyz"]
list_rev.reverse()
print(list_rev)
```

# Tuple

```python
tuple_x = (1,2,45,6.7,"apple","xyz",True,None,12)
print(tuple_x)
tuple_x[2]
tuple_x[-1]
```

```python
tuple_x[3]
```

```python
# indexing -> extract one element at a time
# slicing-> extract multiple element at a time
# datatypes which follows indexing also follow slicing 
list_exam = [10,20,30,40,45]
tuple_y = (10,20,"abc",45,40 ,"xyz")
print(list_exam[1]) # indexing
# slicing syntax
# name_variable[start:end:jump]
# forword slicing 
print(tuple_y[1:4]) # forward direction
```

```python
print(tuple_y[0:2])
```

```python
print(tuple_y[:4]) # by default consider as start is zero
```

```python
print(tuple_y[4:]) # by default consider as end
```

```python
# backword slicing 
# name_variable[start:end:jump]
tuple_y = (11,90.2,"abc",15,40 ,"xyz")
print(tuple_y[-5:-2]) # backrward direction
print(tuple_y[-7:-4])
```

```python
tuple_nested = (11,90.2,3,"abc",(1,11,3,4,"abcde",2.3),15,40,3,"xyz")
# count -> to count the occurance of element
# How many times 11 and 3 appears in the tuple
# it does not count element present inside tuple inside the tuple 
print(tuple_nested.count(11))
print(tuple_nested.count(3))
# index
# to identify index of a particular element
print(tuple_nested.index(90.2)) # 1
# if multiple times on element is present it chosses 1st occurance index of the elemengt
print(tuple_nested.index(3)) # 2 instad of 7
```

# Tuple Reversing 

```python
# how we can reverse a tuple ? is it possible ?
# yes it is possible 
tuple_integer = (1,3,4,2,7,6,5)

# by using slicing method 
print(tuple_integer[::-1])

# 1st convert it to list -> then use reverse method
temp = list(tuple_integer)
temp.reverse()
reversed_tuple = tuple(temp) # convert list -> tuple
print(reversed_tuple)
```

```python
print(tuple_integer[::1])
```

# Sets

declared in {} brackets 
1. unsorted collection of unique element
2. immutable -> add , remove
3. declaration of empty set set_1 = set()
4. does not follow indexing and slicing
5. cannot reverse a set
6. datatypes supported in set -> all primitive and tuple

```python
set_1 = {1,1,2,34,5,5,5,6,6,6,6}
print(set_1)
```

```python
print(set_1)
```

```python
# declaration of empty set 
set_2 = set()
type(set_2)
```

```python
# add and remove 
set_3 = { 1, 2 ,3,4,5,"abc","xyz",34}
set_3.add(50)
set_3.remove(2)
print(set_3)
```

```python
# merging of elements -> union
set_4 = {1,2,3}
set_5 = {4,5,6,7,2,3}
res_union = set_4.union(set_5) # does not considered common element -> as one 2,2, -> 2
print(res_union)

# common elements of sets -> intersection
res_intersection = set_4.intersection(set_5)
print(res_intersection)
```

# Dictionary

1. declared in {}
2. collection of key:value pair
3. does not follow indexing
4. we can declared an empty dictionary
5. we can extract element by key instead of indexing

```python
dict_emp = {}
type(dict_emp)
```

```python
dict_x = {1:"apple", 2:"mango" , 3:"guava" , 4:"grapes" , 5:"litchi"}
print(dict_x)
dict_y = {"fruit_1" : "apple" , "fruit_2" : "mango" , "fruit_3" : "papaya"}
print(dict_y)
```

```python
# extraction 
dict_y["fruit_2"]
```

```python
dict_x[2]
```

```python
# add element -> use key
dict_x = {1:"apple", 2:"mango" , 3:"guava" , 4:"grapes" , 5:"litchi"}
print(dict_x)
dict_x[6] = "orange"
print(dict_x)

# replace element
dict_x[5] = "50"
print(dict_x)
```

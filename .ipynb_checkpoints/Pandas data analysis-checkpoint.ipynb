```python
import pandas as pd
```

```python
# Series -> No column 
S1 = pd.Series(data=[10,20,30,40,50])
# Creating index
S2 = pd.Series(data=[10,20,30], index=["Maths","English","Science"])
S3 = pd.Series(data=[70,80,90], index=["Maths","English","Science"])
print(S2)
print(S3)
```

```python
# To extract data
print(S2["Maths"])
print(S2["English"])
```

```python
print(S1.shape) # No. of rows
print(S1.size) # Total no. of data
print(S1.dtypes) # Data type of a data
```

```python
# Two dimensional things with DataFrame
DF1 = pd.DataFrame(data={"Subjects" : ["Maths","English","Science"] , "Topics" : ["Satastics","Verbal","Species"]})
print(DF1)
```

```python
# dataframe with a pre-defined series
DF2 = pd.DataFrame(data={"sec1":S2,"sec2":S3})
DF2
```
